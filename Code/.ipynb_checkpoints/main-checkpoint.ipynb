## Building the network

### Data load

Functions to process the different Excel sheets in the datasets. *process_funders* returns a dataframe detailing the funders, together with their types, for every trial listed in the "trial funders" sheet.

*process_trials* builds the bipartite network (i.e. a row indicating a connection between a trial and a researcher) from the "Clinical Trials RAW" sheet, including the institutions with which the researcher was affiliated at the time of the trial, and the year.

*process_pub* does a similar work as process_trials, but for the "publications" sheet.

```python
def process_funders(disease="Alzheimers", normalized=True):
    """
    Process funder data for clinical trials, including the list of funding organizations.

    Parameters:
    - disease (str): Name of the disease to load data for.
    - normalized (bool):
        * True: return the fraction of each funder type per trial.
        * False: return the raw counts of each funder type per trial.

    Returns:
    - DataFrame with columns:
        * 'trial'
        * 'funding_orgs'           (list of funding organization names)
        * one column per funder type in ['Nonprofit', 'Education', 'Facility', 
          'Company', 'Government', 'Healthcare'] containing either fractions or counts.
    """
    # 1. Read the funders sheet from the Excel file
    funders = pd.read_excel(
        f"../Data/{disease} Data Delivery.xlsx",
        sheet_name="trial funders",
        header=0
    )

    # 2. Aggregate funding_orgs into a list per trial (unique, preserving order)
    funding_orgs = (
        funders
        .groupby('trial')['funding_orgs']
        .agg(lambda x: list(dict.fromkeys(x.dropna())))
        .rename('funding_orgs')
    )

    # 3. Count occurrences of each funder type per trial
    counts = (
        funders
        .groupby(['trial', 'funder_type'])
        .size()
        .unstack(fill_value=0)
    )

    # 4. Ensure all possible funder type columns are present
    all_types = ['Nonprofit', 'Education', 'Facility', 'Company', 'Government', 'Healthcare']
    counts = counts.reindex(columns=all_types, fill_value=0)

    # 5. Compute fractions or keep raw counts
    if normalized:
        result = counts.div(counts.sum(axis=1), axis=0)
    else:
        result = counts

    # 6. Reset index to turn 'trial' into a column
    result = result.reset_index()

    # 7. Merge the funding_orgs list into the result
    final_df = result.merge(funding_orgs.reset_index(), on='trial', how='left')

    return final_df
```

```python
def process_trials(disease="Alzheimers"):
    """
    Read the clinical-trials Excel and return a flat DataFrame with one row per investigator.

    Parameters:
    - excel_path (str): path to the .xlsx file
    - sheet_name (str): name of the sheet containing the trials data

    Returns:
    - pd.DataFrame with columns:
        * 'trial'
        * 'start_year'
        * 'researcher_name'          (e.g. "Richard T Davey")
        * 'research_organizations'   (list of affiliation strings)
    """
    # 1. Read the sheet
    df = pd.read_excel(
        f"../Data/{disease} Data Delivery.xlsx",
        sheet_name="Clinical Trials RAW",
        header=0
    )
    df.drop_duplicates(subset="trial")

    # 2. Parse the JSON in the 'investigators' column (which contains a top-level "investigators" key)
    def _parse_invs(cell):
        try:
            obj = json.loads(cell)
            return obj.get("investigators", [])
        except Exception:
            return []

    df["investigators"] = df["investigators"].astype(str).apply(_parse_invs)

    # 3. Explode so each investigator gets its own row
    df = df.explode("investigators").reset_index(drop=True)

    # 4. Extract name and affiliations
    def _make_name(inv):
        if not isinstance(inv, dict):
            return None
        first = inv.get("first_name") or ""
        last  = inv.get("last_name")  or ""
        name = f"{first} {last}".strip()
        return name or None

    def _extract_orgs(inv):
        if not isinstance(inv, dict):
            return []
        affs = inv.get("affiliations", [])
        # pull out the "affiliation" field from each entry
        orgs = [a.get("affiliation") for a in affs if isinstance(a, dict) and "affiliation" in a]
        # deduplicate while preserving order
        seen = set()
        unique = []
        for o in orgs:
            if o not in seen:
                seen.add(o)
                unique.append(o)
        return unique

    df["researcher_name"]        = df["investigators"].apply(_make_name)
    df["research_organizations"] = df["investigators"].apply(_extract_orgs)

    # 5. Select only the desired columns
    return df[["trial", "start_year", "researcher_name", "research_organizations"]].dropna().drop_duplicates(subset=["trial", "researcher_name"])
```

```python
def process_pub(disease="Alzheimers"):
    """
    Load and process publication data for a specified disease.

    This function reads the ‚Äúpublications‚Äù and ‚ÄúClinical Trials Expanded‚Äù sheets
    from the Excel file corresponding to the given disease. It groups publications
    by trial and researcher, aggregates each researcher‚Äôs unique organizations 
    into a list, and merges the result with the trial start year. Finally, it drops
    any exact duplicates of (trial, researcher_name).

    Parameters
    ----------
    disease : str, optional
        Name of the disease to load data for (must match the filename prefix in
        "../Data/{disease} Data Delivery.xlsx"), by default "Alzheimers".

    Returns
    -------
    pandas.DataFrame
        DataFrame with the following columns:
        - 'trial': identifier of the clinical trial.
        - 'researcher_name': name of the researcher.
        - 'research_organizations': list of unique organizations for that researcher.
        - 'start_year': start year of the clinical trial.
    """
    # Read publications sheet
    publications = pd.read_excel(
        f"../Data/{disease} Data Delivery.xlsx",
        sheet_name="publications",
        header=0
    )

    # Read clinical trials sheet and keep only trial and start_year
    clinical_trials = pd.read_excel(
        f"../Data/{disease} Data Delivery.xlsx",
        sheet_name="Clinical Trials Expanded",
        header=0
    )
    df_clinical = clinical_trials[['trial', 'start_year']]

    # Group by trial and researcher_name, aggregate unique organizations into a list,
    # and name the aggregated column 'research_organizations'
    df_pub = (
        publications
        .groupby(['trial', 'researcher_name'])['research_organization']
        .agg(lambda orgs: list(set(orgs)))
        .reset_index(name='research_organizations')
    )

    # Merge with clinical trials to bring in start_year
    df_pub = pd.merge(df_pub, df_clinical, on='trial', how='left')

    # Drop any exact duplicates of (trial, researcher_name)
    return df_pub.drop_duplicates(subset=["trial", "researcher_name"])
```

### Bipartite network adjacency lists

```python
def funders_adjacency(network, funder):
    """
    Merge network with funders on 'trial', flag missing funding data, and replace NaNs with 0.
    """
    network_H = pd.merge(network, funder, on="trial", how="left")
    fund_cols = ['Nonprofit', 'Education', 'Facility', 'Company', 'Government', 'Healthcare']

    network_H['no_funding_data'] = network_H[fund_cols].isna().any(axis=1).astype(int)
    network_H[fund_cols] = network_H[fund_cols].fillna(0)

    return network_H
```

```python
def bipartite_summaries(network_H):
    """
    Compute summary statistics for researchers and trials in a bipartite network with funding data.

    This function returns two DataFrames:
      1. researchers_summary: grouped by 'researcher_name', containing:
         - degree: number of trials each researcher participates in
         - mean funding proportions for each funder type ('Nonprofit', 'Education', 'Facility',
           'Company', 'Government', 'Healthcare')
         - mean of 'no_funding_data' indicator
      2. trials_summary: grouped by 'trial', containing:
         - degree: number of researchers involved in each trial
         - mean funding proportions for each funder type
         - mean of 'no_funding_data' indicator

    Parameters
    ----------
    network_H : pandas.DataFrame
        Edge‚Äêlist DataFrame of the bipartite network. Must include columns:
        'researcher_name', 'trial', 'Nonprofit', 'Education', 'Facility',
        'Company', 'Government', 'Healthcare', 'no_funding_data'.

    Returns
    -------
    researchers_summary : pandas.DataFrame
        Summary per researcher with columns ['researcher_name', 'degree',
        'Nonprofit', 'Education', 'Facility', 'Company', 'Government',
        'Healthcare', 'no_funding_data'].
    trials_summary : pandas.DataFrame
        Summary per trial with columns ['trial', 'degree', 'Nonprofit',
        'Education', 'Facility', 'Company', 'Government', 'Healthcare',
        'no_funding_data'].
    """
    researchers_summary = (
        network_H
        .groupby('researcher_name')
        .agg(
            degree=('researcher_name', 'size'),
            Nonprofit=('Nonprofit', 'mean'),
            Education=('Education', 'mean'),
            Facility=('Facility', 'mean'),
            Company=('Company', 'mean'),
            Government=('Government', 'mean'),
            Healthcare=('Healthcare', 'mean'),
            no_funding_data=('no_funding_data', 'mean')
        )
        .reset_index()
    )

    trials_summary = (
        network_H
        .groupby('trial')
        .agg(
            degree=('trial', 'size'),
            Nonprofit=('Nonprofit', 'mean'),
            Education=('Education', 'mean'),
            Facility=('Facility', 'mean'),
            Company=('Company', 'mean'),
            Government=('Government', 'mean'),
            Healthcare=('Healthcare', 'mean'),
            no_funding_data=('no_funding_data', 'mean')
        )
        .reset_index()
    )

    return researchers_summary, trials_summary
```

```python
def funding_degree(network, normalized=True, n_bins=None):
    """
    Compute funding distribution by researcher degree, optionally binning degrees.

    Parameters:
    - network (pd.DataFrame): must contain a 'degree' column and these funder columns:
        ['Nonprofit', 'Education', 'Facility', 'Company', 'Government', 'Healthcare', 'no_funding_data']
    - normalized (bool):
        * True: sum the counts in each degree/bin and then divide by the total counts 
                (so each bin‚Äôs values sum to 1).
        * False: compute the average fraction of each funder type (mean of the 0/1 indicators) 
                per degree/bin (original behavior).
    - n_bins (int or None): number of degree intervals to group into; if None, group by each distinct degree.

    Returns:
    - dict[str, list[float]]: mapping each degree-range label (e.g. "1-1" or "1-10") 
      to a list of values for each funder column in the order:
      ['Nonprofit', 'Education', 'Facility', 'Company', 'Government', 'Healthcare', 'no_funding_data']
    """
    fund_cols = [
        'Nonprofit', 'Education', 'Facility',
        'Company', 'Government', 'Healthcare',
        'no_funding_data'
    ]
    degrees = network['degree']

    if n_bins is None:
        # No binning: group by each distinct degree
        if normalized:
            grouped = network.groupby('degree')[fund_cols].sum()
            grouped = grouped.div(grouped.sum(axis=1), axis=0)
        else:
            grouped = network.groupby('degree')[fund_cols].mean()
        # Build dict with labels "d-d"
        return {
            f"{int(deg)}-{int(deg)}": grouped.loc[deg].tolist()
            for deg in grouped.index
        }

    # Binning: cut degrees into n_bins equal-width intervals
    bins = pd.cut(degrees, bins=n_bins)
    # Create labels "left-right" for each interval
    labels = [
        f"{int(interval.left)}-{int(interval.right)}"
        for interval in bins.cat.categories
    ]
    binned = pd.cut(degrees, bins=n_bins, labels=labels, include_lowest=True)

    # Attach bins to a copy of the DataFrame
    df2 = network.copy()
    df2['degree_bin'] = binned

    if normalized:
        summed = df2.groupby('degree_bin')[fund_cols].sum()
        grouped = summed.div(summed.sum(axis=1), axis=0)
    else:
        grouped = df2.groupby('degree_bin')[fund_cols].mean()

    return {
        bin_label: grouped.loc[bin_label].tolist()
        for bin_label in grouped.index
    }


def funding_degree_graph(network, normalized=True, n_bins=None):
    """
    Plot funding distribution by researcher degree (or degree bins) as a stacked bar chart.

    Parameters:
    - network (pd.DataFrame): must contain a 'degree' column and these funder columns:
        ['Nonprofit', 'Education', 'Facility', 'Company', 'Government', 'Healthcare', 'no_funding_data']
    - normalized (bool):
        * True: sum then normalize (each bar sums to 1).
        * False: plot average fractions per degree/bin (original behavior).
    - n_bins (int or None): number of degree intervals to group into; if None, group by each distinct degree.

    Returns:
    - None: displays the matplotlib figure.
    """
    fund_cols = [
        'Nonprofit', 'Education', 'Facility',
        'Company', 'Government', 'Healthcare',
        'no_funding_data'
    ]
    degrees = network['degree']

    if n_bins is None:
        # No binning
        if normalized:
            data = network.groupby('degree')[fund_cols].sum()
            data = data.div(data.sum(axis=1), axis=0)
            xlabel = 'Degree'
            title_suffix = 'Normalized distribution'
        else:
            data = network.groupby('degree')[fund_cols].mean()
            xlabel = 'Degree'
            title_suffix = 'Average distribution'
        x_labels = [f"{int(d)}-{int(d)}" for d in data.index]

    else:
        # Binning
        bins = pd.cut(degrees, bins=n_bins)
        labels = [
            f"{int(interval.left)}-{int(interval.right)}"
            for interval in bins.cat.categories
        ]
        binned = pd.cut(degrees, bins=n_bins, labels=labels, include_lowest=True)
        df2 = network.copy()
        df2['degree_bin'] = binned

        if normalized:
            summed = df2.groupby('degree_bin')[fund_cols].sum()
            data = summed.div(summed.sum(axis=1), axis=0)
            title_suffix = 'Normalized distribution'
        else:
            data = df2.groupby('degree_bin')[fund_cols].mean()
            title_suffix = 'Average distribution'
        x_labels = list(data.index)
        xlabel = 'Degree range'

    bottoms = np.zeros(len(x_labels))
    fig, ax = plt.subplots()
    for col in fund_cols:
        vals = data[col].values
        ax.bar(x_labels, vals, bottom=bottoms, label=col)
        bottoms += vals

    ax.set_xlabel(xlabel)
    ax.set_ylabel('Fraction' if normalized else 'Count')
    ax.set_title(f'Funding Distribution by Degree ({title_suffix})')
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## Temporal networks

```python
def build_bipartite_networks(df):
    """
    Build year-specific and cumulative bipartite networks connecting trials and researchers. Returns a sequence
    of graphs G(t) for all years for which we have data.

    Parameters:
    - df (pd.DataFrame): must contain columns:
        * 'trial' (str)
        * 'researcher_name' (str)
        * 'start_year' (numeric)

    Returns:
    - year_graphs (dict): mapping each year -> bipartite Graph of edges occurring in that year
    - cumulative_graphs (dict): mapping each year -> bipartite Graph of all edges up to and including that year
    """
    # Ensure years are sorted and drop missing
    years = sorted(df['start_year'].dropna().unique())
    
    year_graphs = {}
    cumulative_graphs = {}
    G_cum = nx.Graph()
    
    for year in years:
        # Filter rows for this exact year
        df_year = df[df['start_year'] == year]
        
        # Create a fresh graph for edges in this year
        G_year = nx.Graph()
        
        for trial, researcher in df_year[['trial', 'researcher_name']].itertuples(index=False):
            # Add trial node in set 0
            if not G_year.has_node(trial):
                G_year.add_node(trial, bipartite=0)
            # Add researcher node in set 1
            if not G_year.has_node(researcher):
                G_year.add_node(researcher, bipartite=1)
            # Connect trial <-> researcher
            G_year.add_edge(trial, researcher)
            
            # Also update the cumulative graph
            if not G_cum.has_node(trial):
                G_cum.add_node(trial, bipartite=0)
            if not G_cum.has_node(researcher):
                G_cum.add_node(researcher, bipartite=1)
            G_cum.add_edge(trial, researcher)
        
        year_graphs[year] = G_year
        # Store a copy to freeze the state at this year
        cumulative_graphs[year] = G_cum.copy()
    
    return year_graphs, cumulative_graphs
```

```python
def plot_bipartite_graph(
    G: nx.Graph,
    figsize: tuple = (10, 6),
    top_color: str = "#1f78b4",
    bottom_color: str = "#33a02c",
    node_size: int = 100,
    with_labels: bool = False
):
    """
    Draw a bipartite network using a two‚Äêcolumn layout.

    Parameters:
    - G (nx.Graph): bipartite graph with node attribute 'bipartite' equal to 0 or 1.
    - figsize (tuple): figure size passed to plt.figure.
    - top_color (str): color for the nodes in bipartite set 0 (trials).
    - bottom_color (str): color for the nodes in bipartite set 1 (researchers).
    - node_size (int): size of the node markers.
    - with_labels (bool): whether to draw node labels.
    """
    # 1. Identify the two node sets by the bipartite attribute
    top_nodes    = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 0]
    bottom_nodes = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 1]

    # 2. Compute a bipartite layout: top_nodes on the left, bottom_nodes on the right
    pos = nx.bipartite_layout(G, top_nodes)

    # 3. Create the figure
    plt.figure(figsize=figsize)

    # 4. Draw edges
    nx.draw_networkx_edges(G, pos, alpha=0.3)

    # 5. Draw the two node sets with different colors and shapes
    nx.draw_networkx_nodes(
        G, pos,
        nodelist=top_nodes,
        node_shape='s',
        node_color=top_color,
        node_size=node_size,
        label='Trials'
    )
    nx.draw_networkx_nodes(
        G, pos,
        nodelist=bottom_nodes,
        node_shape='o',
        node_color=bottom_color,
        node_size=node_size,
        label='Researchers'
    )

    # 6. Optionally draw labels
    if with_labels:
        nx.draw_networkx_labels(G, pos, font_size=8)

    # 7. Final touches
    plt.axis('off')
    plt.legend(scatterpoints=1, loc='upper right')
    plt.title("Bipartite Network: Trials ‚Üî Researchers")
    plt.tight_layout()
    plt.show()
```

```python
def bipartite_researcher_clustering(G, mode='dot', return_avg=True):
    """
    Calculate the bipartite clustering coefficient for researcher nodes only.

    Parameters:
    - G (nx.Graph): bipartite graph with node attribute 'bipartite' equal to 0 or 1.
    - mode (str): one of "dot", "ops", or "jaccard" (see networkx docs for definitions).
    - return_avg (bool): if True, also return the average clustering across researcher nodes.

    Returns:
    - clustering (dict): mapping each researcher node ‚Üí its bipartite clustering coefficient.
    - avg_clustering (float, optional): average of the coefficients (only if return_avg=True).
    """
    # 1. Select researcher nodes (bipartite set = 1)
    researchers = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 1]

    # 2. Compute local clustering for each researcher
    clustering = bipartite.clustering(G, nodes=researchers, mode=mode)

    if return_avg:
        avg = sum(clustering.values()) / len(clustering) if clustering else 0.0
        return clustering, avg
    else:
        return clustering
```

#### Degree distribution

```python
def researcher_degree_distribution(G):
    """
    Compute the degree distribution for researcher nodes in a bipartite graph.

    Parameters:
    - G (nx.Graph): a bipartite graph where researcher nodes have attribute bipartite=1.

    Returns:
    - dict[int, float]: mapping each degree k to its probability P(k),
      i.e., the fraction of researcher nodes having degree k.
    """
    # 1. Identify researcher nodes (bipartite set = 1)
    researchers = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 1]

    # 2. Get their degrees
    degrees = [G.degree(n) for n in researchers]

    # 3. Count occurrences of each degree
    counts = Counter(degrees)
    total = len(researchers)

    # 4. Convert counts to probabilities
    return {k: v / total for k, v in counts.items()}
```

```python
def bins(values, B):
    """
    Compute logarithmic bins from an array of values.

    Parameters:
    - values (array-like): numeric values to determine min and max.
    - B (int): number of bins.

    Returns:
    - bins_linspace (np.ndarray of shape (B,2)): linear-space interval edges [(lo, hi), ...]
    """
    vals = np.array(values, dtype=float)
    x_min = 1 if vals.min() == 0 else vals.min()
    x_max = vals.max()
    log_min, log_max = np.log(x_min), np.log(x_max)
    edges = np.exp(np.linspace(log_min, log_max, B + 1))
    # pair up edges into (lo, hi)
    bins_linspace = np.vstack([edges[:-1], edges[1:]]).T
    return bins_linspace

def log_binning(value_freq, B):
    """
    Perform log-binning on a distribution given as {value: frequency},
    returning bin centers and the probability mass in each bin.

    Parameters:
    - value_freq (dict): mapping each data value to its frequency (count).
    - B (int): number of log-spaced bins.

    Returns:
    - X_binned (list): geometric mean of each bin interval.
    - Y_binned (list): probability mass per bin (sum of freqs in bin / total count).
    """
    # 1. Extract unique values and total count
    values = np.array(list(value_freq.keys()), dtype=float)
    total_count = sum(value_freq.values())

    # 2. Compute B log-spaced bins
    bins_lin = bins(values, B)

    # 3. Allocate counts into bins
    bin_counts = np.zeros(B, dtype=float)
    for val, freq in value_freq.items():
        for i, (lo, hi) in enumerate(bins_lin):
            if lo <= val < hi or (i == B-1 and val == hi):
                bin_counts[i] += freq
                break

    # 4. Compute probability mass in each bin
    Y_binned = (bin_counts / total_count).tolist()

    # 5. Compute bin centers (geometric mean)
    X_binned = [np.sqrt(lo * hi) for lo, hi in bins_lin]

    return X_binned, Y_binned
```

#### 4-cycles

```python
def count_4_cycles(G):
    """
    Count the number of unique 4-cycles (squares) in a bipartite graph.

    A 4-cycle in a bipartite graph corresponds to two nodes in set U
    (bipartite=0) sharing two distinct neighbors in set V (bipartite=1).

    Parameters:
    - G (nx.Graph): bipartite graph with node attribute 'bipartite' = 0 or 1.

    Returns:
    - int: total number of 4-cycles.
    """
    # Identify the ‚Äútrial‚Äù side (bipartite=0)
    U = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 0]
    total = 0
    # For each unordered pair of U-nodes, count pairs of common neighbors in V
    for i, u1 in enumerate(U):
        nbrs1 = set(G.neighbors(u1))
        for u2 in U[i+1:]:
            common = nbrs1 & set(G.neighbors(u2))
            k = len(common)
            if k >= 2:
                # each pair of common neighbors defines one unique square
                total += comb(k, 2)
    return total


def count_new_squares(cumulative_graphs):
    """
    For each year, compute how many new 4-cycles appear in the cumulative graph
    compared to the previous year.

    Parameters:
    - cumulative_graphs (dict[int, nx.Graph]):
        mapping year -> cumulative bipartite graph up to that year.

    Returns:
    - dict[int, int]: mapping year t -> number of new squares formed between t-1 and t.
      (First year is skipped since there is no t-1.)
    """
    years = sorted(cumulative_graphs)
    # Pre-compute total 4-cycles per year
    cycles_per_year = {
        year: count_4_cycles(cumulative_graphs[year])
        for year in years
    }
    # Compute year-over-year increases
    new_squares = {}
    for prev, curr in zip(years, years[1:]):
        new_squares[curr] = cycles_per_year[curr] - cycles_per_year[prev]
    return new_squares
```

#### Effective size

```python
def effective_size_bipartite(B, mode=1):
    """
    B: bipartite NetworkX graph with node attribute 'bipartite' in {0,1}
    mode: which node set to project (0 or 1)
    returns: dict mapping node -> effective size S_i
    """
    # 1. Weighted one-mode projection onto the chosen set
    nodes_to_project = [n for n, d in B.nodes(data=True) if d.get('bipartite') == mode]
    P = nx.bipartite.weighted_projected_graph(B, nodes_to_project)
    
    S = {}
    for i in P:
        neighbors_i = list(P[i])
        # if isolated in projection, effective size = 0
        if not neighbors_i:
            S[i] = 0
            continue
        
        # total tie-strength from i
        tot = sum(P[i][j]['weight'] for j in neighbors_i)
        if tot == 0:
            S[i] = 0
            continue
        
        # proportions p_ij
        p = {j: P[i][j]['weight'] / tot for j in neighbors_i}
        
        # compute effective size for node i
        S_i = 0
        for j in neighbors_i:
            neighbors_j = list(P[j])
            denom_j = sum(P[j][q]['weight'] for q in neighbors_j)
            if denom_j == 0:
                continue
            
            # redundancy r_ij: sum over common neighbors only
            r_ij = sum(
                p[q] * (P[j][q]['weight'] / denom_j)
                for q in neighbors_i
                if q in P[j]
            )
            S_i += 1 - r_ij
        
        S[i] = S_i
    
    return S
```

### Network mobility

```python
def researcher_degree_increase_probability(cumulative_graphs):
    """
    Compute, for each researcher degree k, the probability that
    in the next year the researcher‚Äôs cumulative degree increases by at least one.

    Parameters:
    - cumulative_graphs (dict[int, nx.Graph]):
        mapping year -> cumulative bipartite graph up to that year.
        Researcher nodes must have node attribute bipartite == 1.

    Returns:
    - dict[int, float]: mapping each starting degree k ‚Üí
        P(degree_{t+1} ‚â• k+1 | degree_t = k), estimated over all researchers.
    """
    years = sorted(cumulative_graphs)
    total_transitions = defaultdict(int)
    successful_transitions = defaultdict(int)

    for year_t, year_t1 in zip(years, years[1:]):
        G_t  = cumulative_graphs[year_t]
        G_t1 = cumulative_graphs[year_t1]

        # only researcher nodes (bipartite = 1) present in year_t
        researchers = [n for n, d in G_t.nodes(data=True) if d.get('bipartite') == 1]

        for r in researchers:
            k_t = G_t.degree(r)
            # if researcher appears in G_t1, use its new degree; otherwise it stays k_t
            k_t1 = G_t1.degree(r) if G_t1.has_node(r) else k_t

            total_transitions[k_t] += 1
            if k_t1 >= k_t + 1:
                successful_transitions[k_t] += 1

    # build probability dict
    prob_increase = {
        k: successful_transitions[k] / total_transitions[k]
        for k in total_transitions
    }
    return prob_increase
```

```python
def fit_degree_increase_lm(degree_probs):
    """
    Fit a linear model P(increase) = a + b * degree
    from a mapping of degree to probability of increase.

    Parameters
    ----------
    degree_probs : dict[int, float]
        Dictionary where each key is a node degree and each value is the
        corresponding probability of an increase.

    Returns
    -------
    sklearn_model : sklearn.linear_model.LinearRegression
        The fitted LinearRegression instance.
    coef : float
        The slope coefficient (b).
    intercept : float
        The intercept term (a).
    r2 : float
        R-squared value of the fit.
    sm_result : statsmodels.regression.linear_model.RegressionResultsWrapper
        The result object from a statsmodels OLS fit, including diagnostics.
    """
    # Convert dictionary to DataFrame
    df = pd.DataFrame(
        list(degree_probs.items()),
        columns=['degree', 'prob_increase']
    )

    # Prepare feature matrix X and target vector y
    X = df[['degree']].values         # shape (n_samples, 1)
    y = df['prob_increase'].values    # shape (n_samples,)

    # Fit linear model with scikit-learn
    lm = LinearRegression()
    lm.fit(X, y)
    coef = lm.coef_[0]        # slope b
    intercept = lm.intercept_ # intercept a
    r2 = lm.score(X, y)       # R¬≤ of the fit

    # Fit with statsmodels for statistical diagnostics
    X_sm = sm.add_constant(df['degree'])  # add intercept column
    sm_model = sm.OLS(df['prob_increase'], X_sm).fit()

    return lm, coef, intercept, r2, sm_model
```

### Institution collaboration matrix

```python
def build_bipartite_networks(df):
    """
    Build year-specific and cumulative bipartite networks connecting trials and researchers,
    and attach each researcher node the list of organizations they belong to.

    Parameters:
    - df (pd.DataFrame): must contain columns:
        * 'trial' (str)
        * 'researcher_name' (str)
        * 'research_organizations' (list of str)
        * 'start_year' (numeric)

    Returns:
    - year_graphs (dict[int, nx.Graph]): mapping each year -> bipartite Graph of edges in that year
    - cumulative_graphs (dict[int, nx.Graph]): mapping each year -> bipartite Graph of all edges up to that year
    """
    years = sorted(df['start_year'].dropna().unique())
    year_graphs = {}
    cumulative_graphs = {}
    G_cum = nx.Graph()

    # keep track of cumulative orgs per researcher
    cumulative_orgs = {}

    for year in years:
        df_year = df[df['start_year'] == year]
        G_year = nx.Graph()
        # track orgs for this year only
        year_orgs = {}

        for trial, researcher, orgs in df_year[['trial', 'researcher_name', 'research_organizations']].itertuples(index=False):
            # ensure orgs is a list
            org_list = orgs if isinstance(orgs, list) else []
            # update per-year and cumulative org sets
            year_orgs.setdefault(researcher, set()).update(org_list)
            cumulative_orgs.setdefault(researcher, set()).update(org_list)

            # ---- Year graph ----
            if not G_year.has_node(trial):
                G_year.add_node(trial, bipartite=0)
            if not G_year.has_node(researcher):
                G_year.add_node(
                    researcher,
                    bipartite=1,
                    organizations=list(year_orgs[researcher])
                )
            else:
                # refresh the organizations attribute
                G_year.nodes[researcher]['organizations'] = list(year_orgs[researcher])
            G_year.add_edge(trial, researcher)

            # ---- Cumulative graph ----
            if not G_cum.has_node(trial):
                G_cum.add_node(trial, bipartite=0)
            if not G_cum.has_node(researcher):
                G_cum.add_node(
                    researcher,
                    bipartite=1,
                    organizations=list(cumulative_orgs[researcher])
                )
            else:
                G_cum.nodes[researcher]['organizations'] = list(cumulative_orgs[researcher])
            G_cum.add_edge(trial, researcher)

        year_graphs[year] = G_year
        cumulative_graphs[year] = G_cum.copy()

    return year_graphs, cumulative_graphs
```

```python
def build_institution_collab_matrices(cumulative_graphs):
    """
    For each year‚Äôs cumulative bipartite network, compute the
    institution‚Äìinstitution collaboration counts based on shared trials.

    Parameters:
    - cumulative_graphs (dict[int, nx.Graph]):
        mapping year -> bipartite graph up to that year. Researcher
        nodes must have a node attribute 'bipartite' == 1 and an
        attribute 'organizations' which is a list of institution strings.

    Returns:
    - dict[int, pd.DataFrame]: mapping each year -> symmetric DataFrame
      whose rows and columns are institution names, and whose (i, j)
      entry is the number of collaborations between institution i and j
      (i ‚â† j) up to that year. Years are ordered from most recent to oldest.
    """
    matrices = {}
    # iterate from most recent year down
    for year in sorted(cumulative_graphs.keys(), reverse=True):
        G = cumulative_graphs[year]
        # collect all collaboration pairs
        collab_counts = Counter()
        
        # iterate over each trial node
        trials = [n for n, d in G.nodes(data=True) if d.get('bipartite') == 0]
        for trial in trials:
            # get all researchers on this trial
            researchers = [nbr for nbr in G.neighbors(trial)
                           if G.nodes[nbr].get('bipartite') == 1]
            # for each unordered pair of researchers
            for r1, r2 in combinations(researchers, 2):
                orgs1 = G.nodes[r1].get('organizations', []) or []
                orgs2 = G.nodes[r2].get('organizations', []) or []
                # count cross-institution collaborations
                for inst1 in orgs1:
                    for inst2 in orgs2:
                        if inst1 == inst2:
                            continue
                        # use sorted tuple to keep it unordered
                        key = tuple(sorted((inst1, inst2)))
                        collab_counts[key] += 1
        
        # build the square DataFrame
        insts = sorted({i for pair in collab_counts for i in pair})
        mat = pd.DataFrame(0, index=insts, columns=insts, dtype=int)
        for (i, j), cnt in collab_counts.items():
            mat.at[i, j] = cnt
            mat.at[j, i] = cnt
        
        matrices[year] = mat
    
    return matrices
```

```python
def visualize_institution_matrix(mat,
                                 figsize=(8,8)):
    """
    Visualize a large sparse institution‚Äìinstitution collaboration matrix.

    Parameters:
    - mat (pd.DataFrame): symmetric adjacency matrix (zero = no collab).
    - top_n (int or None):
        * None: show sparsity pattern (plt.spy).
        * int: take top_n institutions by total collaborations, 
               build a weighted graph and plot it.
    - threshold (int): when top_n is set, drop edges with weight < threshold.
    - figsize (tuple): figure size for the plot.
    - seed (int): random seed for layout reproducibility.
    """
    # ---- Sparsity pattern ----
    plt.figure(figsize=figsize)
    plt.spy(mat, markersize=1, precision=0.1)
    plt.title("Sparsity pattern of institution collaborations")
    plt.xlabel("Institution index")
    plt.ylabel("Institution index")
    plt.tight_layout()
    plt.show()
    
```

```python
def visualize_institution_matrix_colored(
    mat,
    figsize=(10,8),
    cmap='viridis',
    markersize=1,
    vmax=None    # si no es None, todos los valores > vmax se mostrar√°n como vmax
):
    """
    Visualize a sparse institution‚Äìinstitution collaboration matrix,
    coloring each non-zero entry by its weight, con opci√≥n de capar el m√°ximo.

    Parameters:
    - mat (pd.DataFrame): symmetric adjacency matrix (zero = no collab).
    - figsize (tuple): figure size.
    - cmap (str): name of the matplotlib colormap to use.
    - markersize (float): size of the scatter markers.
    - vmax (float or None): m√°ximo valor de color; todo valor > vmax se muestra con el color de vmax.
    """
    # 1. Extraer √≠ndices de fila, columna y valores no-ceros
    vals_mat = mat.values
    rows, cols = np.nonzero(vals_mat)
    vals       = vals_mat[rows, cols]

    # 2. Si se pide capar, clippear los valores
    if vmax is not None:
        vals_to_color = np.clip(vals, None, vmax)
        norm = plt.Normalize(vmin=0, vmax=vmax)
    else:
        vals_to_color = vals
        norm = None

    # 3. Dibujar con scatter
    fig, ax = plt.subplots(figsize=figsize)
    sc = ax.scatter(
        cols,
        rows,
        c=vals_to_color,
        cmap=cmap,
        s=markersize,
        marker='s',
        norm=norm
    )
    ax.invert_yaxis()
    ax.set_xlabel("Institution index")
    ax.set_ylabel("Institution index")
    ax.set_title("Institution Collaborations (colored by count)")

    # 4. Barra de color
    cbar = plt.colorbar(sc, ax=ax, label='Collaboration count')

    plt.tight_layout()
    plt.show()
```

### New collaborations analysis

```python
def count_new_researcher_connections(cumulative_graphs):
    """
    For each year, count how many new researcher‚Äìresearcher connections
    appear in the one‚Äêmode projection (shared trial) that were not present
    in the previous year's cumulative network, considering only pairs
    where both researchers already existed in the prior year.

    Parameters:
    - cumulative_graphs (dict[int, nx.Graph]):
        mapping year -> cumulative bipartite graph up to that year.
        Researcher nodes must have node attribute 'bipartite' == 1.

    Returns:
    - dict[int, int]: mapping each year t -> number of new researcher‚Äìresearcher connections
      formed between year t-1 and t.
      (The first year is omitted, since there's no prior graph to compare.)
    """
    years = sorted(cumulative_graphs)
    new_conn_counts = {}

    # Build projection for first year
    first = years[0]
    G_prev = cumulative_graphs[first]
    researchers_prev = [n for n, d in G_prev.nodes(data=True) if d.get('bipartite') == 1]
    P_prev = bipartite.weighted_projected_graph(G_prev, researchers_prev)

    # Iterate year by year
    for year in years[1:]:
        G_curr = cumulative_graphs[year]
        # researchers in current cumulative graph
        researchers_curr = [n for n, d in G_curr.nodes(data=True) if d.get('bipartite') == 1]
        P_curr = bipartite.weighted_projected_graph(G_curr, researchers_curr)

        count = 0
        # Look for edges in P_curr that are new relative to P_prev
        for u, v, data in P_curr.edges(data=True):
            # only count if both existed previously
            if u in researchers_prev and v in researchers_prev:
                weight_prev = P_prev[u][v]['weight'] if P_prev.has_edge(u, v) else 0
                weight_curr = data.get('weight', 0)
                # new connection if weight_prev == 0 and now weight_curr > 0
                if weight_prev == 0 and weight_curr > 0:
                    count += 1

        new_conn_counts[year] = count

        # update for next iteration
        G_prev = G_curr
        P_prev = P_curr
        researchers_prev = researchers_curr

    return new_conn_counts
```
